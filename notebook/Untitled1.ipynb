```python
import seaborn as sns
import sklearn
import pandas as pd
import numpy as np

from sklearn.preprocessing import OrdinalEncoder
from sklearn.ensemble import GradientBoostingRegressor

import joblib
import pickle
```

```python
df = sns.load_dataset('tips')
```

```python
oe = OrdinalEncoder()

categorical_col = df.select_dtypes(exclude = 'number').columns
new_df = oe.fit_transform(df[categorical_col])
```

```python
df = df.drop(['sex', 'smoker', 'day', 'time'], axis = 1)
```

```python
new_df = pd.DataFrame(new_df, columns = categorical_col)
df = pd.concat([df, new_df], axis = 1)
```

```python
df['day'].value_counts()
```

```python
# CORRECT: Save the actual encoder object!
with open('Tips_Encoder.pkl', 'wb') as file:
    pickle.dump(oe, file)
```

```python
x = df.drop(['tip'], axis = 1)
y = df['tip']
model = GradientBoostingRegressor()
model.fit(x, y)
```

```python
# CORRECT: Save the actual encoder object!
with open('tips_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```
